```python
from tensorflow import keras
from numpy import random
from Brett import conv_model
```

### Hyperparameters

```python
H = {
    'input_sigs': ['II', 'V', 'PLETH', 'RESP'],
    'layer_count_a': 3,
    'layer_count_b': 5,
    'window_size': 512,
    'activation': 'relu',
    'dropout': 0.2,
    'filter_count': 128,
    'dense_units': 64,
    'kernel_size_a': 16,
    'stride_a': 4,
    'kernel_size_b': 4,
    'batch_size': 32,
    'batch_buffer_size': 10,
    'optimizer': 'adam',
    'learning_rate': 3e-4,
}
```

### Build Model

```python
model = conv_model.build(H)
```

### Test Batch

```python
x = random.rand(H['batch_size'], H['window_size'], len(H['input_sigs']))
y = random.rand(H['batch_size'], 2)
model.test_on_batch(x, y)
```

### Create Vector Graphic (requires GraphViz installed)

```python
model_dot = keras.utils.model_to_dot(model, dpi=60, show_shapes=True, show_layer_names=False)
model_svg = model_dot.create(prog='dot', format='svg')
with open('Brett_first_model.svg', 'wb') as f: f.write(model_svg)
```
